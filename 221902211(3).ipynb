```python
#Sum of digits in a number
n = int(input("Put a number: "))
sum = 0;
while n>0:
  remainder=n%10
  n//=10
  sum+=remainder
print (sum)
```

```python
#Reverse a List
A = [36, 69, 24, 32, 45]
print(A)
i=0
while i<len(A)/2:
  A[i], A[len(A)-i-1] = A[len(A)-i-1], A[i]
  i+=1
print(A)
```

```python
#Finding the maximum number
A = [36, 69, 24, 32, 45, 93]
max = A[0]
i = 0
while i<len(A):
  if max < A[i]:
    max = A[i]
    i+= 1
print(max)
```
