```python
product = {
    'Laptop': 10000,
    'smart phone': 5000,
    'tab': 40000,
    'tv': 1200
}

sum = 0
for value in product.values():
  sum = sum + value
print(sum)
```

```python
dic = {'a': 4, 'b': 1, 'c': 6, 'd': 2, 'e': 3}
newdic = {k:v*5 for (k,v) in dic.items()}
print(newdic)

newdic = {k:v for (k,v) in dic.items() if 'a' in k}
print(newdic)
```

```python
color = {'red': '34h',
         'blue': '3f3',
         'yellow': '45y'}
nwcolor = {K.upper(): V.upper() for K, V in color.items()}
print(nwcolor)
```
