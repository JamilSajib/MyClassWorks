For loop in python

```python
for i in range(1,10):
  print('Value of i is:', i)
```

```python
for i in range(1, 50):
  print('The original Values in the array: ', i)
  if(i%2==0):
    print('The even numbers in the array: ', i)
```

```python
for i in range(1,5):
  if(i>2):
   break
  print(i)
```

```python
House = ['Living_Room','Parking_Space','Garden']
for h in House:
  print(h)
```

```python
Alphabet = {'A', 'B', 'C'}
for i,letter in enumerate(Alphabet):
  print(i,letter)
```
